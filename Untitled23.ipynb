```python
data = data.query('(con_treat == "control" and page == "old_page") or (con_treat == "treatment" and page == "new_page")')
```

```python
a_group = data.query('con_treat == "control"')
b_group = data.query('con_treat == "treatment"')
```

```python
b_group.describe()
```

```python
a_group.describe()
```

```python
n1 = a_group.converted.value_counts().sum()
n2 = b_group.converted.value_counts().sum()
k1 = a_group.converted.value_counts()[1]
k2 = b_group.converted.value_counts()[1]
```

```python
from statsmodels.stats import proportion

chisq, pvalue, table = proportion.proportions_chisquare(np.array([k1, k2]),
                                                   np.array([n1, n2]))

print('Results are ','chisq =%.3f, pvalue = %.3f'%(chisq, pvalue))
```

```python
z_score, z_pvalue = proportion.proportions_ztest(np.array([k1, k2]),
                                                   np.array([n1, n2]))
print('Results are ','z_score =%.3f, pvalue = %.3f'%(z_score, z_pvalue))
```
